```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
dat : np.array = np.load('../TimingData/signal.npy').flatten()
time : np.array = np.load('../TimingData/timing.npy').flatten()
x   = np.arange(len(dat))
l   = -1
m   = 24
u   : np.array = np.roll(np.load('../TimingData/bits.npy').flatten(),shift=1)

fig, ax1 = plt.subplots()
# Plot dat on the left y-axis
LL = dat[:l*m]-u[:l*m]
ax1.plot(x[:l*m], LL, color='tab:blue', label='Signal')
ax1.set_ylabel(r'Signal Error [$\mathit{bit}_{\mathit{actual}} - \mathit{bit}_{\mathit{expected}}$]', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis sharing the same x-axis
ax2 = ax1.twinx()
ax2.plot(x[:l*m], time[:l*m], color='tab:red', label=r'Time $\mu s$',alpha=0.5)
ax2.set_ylabel(r'time [$\mu s$]', color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

ax1.set_xlabel("Frame number")
ax2.set_xlabel("Frame number")
plt.title('System timing diagram')
plt.show()
```

```python
dat : np.array = np.load('../TimingData/signal.npy').flatten()
time : np.array = np.load('../TimingData/vs.npy').flatten()
x   = np.arange(len(dat))
s   = 0
m   = -1
u   : np.array = np.roll(np.load('../TimingData/bits.npy').flatten(),shift=0)

fig, ax1 = plt.subplots()
# Plot dat on the left y-axis
LL = dat[s:m]-u[s:m]
ax1.plot(x[s:m], LL, color='tab:blue', label='Signal')
ax1.set_ylabel(r'Signal Error [$\mathit{bit}_{\mathit{actual}} - \mathit{bit}_{\mathit{expected}}$]', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis sharing the same x-axis
ax2 = ax1.twinx()
ax2.plot(x[s:m], time[s:m], color='tab:red', label=r'Time $\mu s$',alpha=0.5)
ax2.set_ylabel(r'VSync Difference [$\Delta$frame]', color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

ax1.set_xlabel("Frame number")
ax2.set_xlabel("Frame number")
plt.title('System timing diagram')
plt.show()
```
