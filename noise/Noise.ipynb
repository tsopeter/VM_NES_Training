```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
imgs = []
for i in range(10):
    imgs.append(cv2.imread(f"00{i}.bmp",cv2.IMREAD_GRAYSCALE))
imgs = np.array(imgs)

img = np.mean(imgs, axis=0)

# Divide into 16 quadrants of equal size
H, W = img.shape

qH = H // 4
qW = qH

# note qH and qW are the same, this means we have "empty" space
# for qW
quads = []
s = (W - qW * 4) // 3
for i in range(4):
    for j in range(4):
        quads.append(img[i*qH:(i+1)*qH, j*(qW+s):j*(qW+s)+qW])

# Print it out in a 4x4 grid
plt.figure(figsize=(8,8))
for i in range(4):
    for j in range(4):
        plt.subplot(4, 4, i*4+j+1)
        plt.imshow(quads[i*4+j], cmap='gray')
        plt.title("Sum: {:.1f}".format(np.sum(quads[i*4+j])))
        plt.axis('off')
plt.show()
# print out the values as a array
print([int(np.sum(q)) for q in quads[:10]])
```

```python
def get_quads(img):
    quads = []
    s = (W - qW * 4) // 3
    for i in range(4):
        for j in range(4):
            quads.append(img[i*qH:(i+1)*qH, j*(qW+s):j*(qW+s)+qW])
    return quads

imgs = []
for i in range(10):
    imgs.append(cv2.imread(f"00{i}.bmp",cv2.IMREAD_GRAYSCALE))

qt = [
    get_quads(i) for i in imgs
]
qt = np.array(qt)
qt = np.sum(qt, axis=(2,3))
```

```python
plt.plot(100*qt/(255*(60*60)))
plt.show()
```
