```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from PythonTools.data_reader import read_file
```

```python
fashion_data = {
    'Categorical' : 
    {
        'Data' : read_file('2026_01_27/16/categorical/fashion/008_59_b_lr_0.15_sz_10'),
        'Color' : 'blue'
    },
    'Normal'      : 
    {
        'Data' : read_file('2026_01_27/16/normal/fashion/002_59_b_lr_0.15_sz_10'),
        'Color' : 'orange'
    }
}

mnist_data = {
    'Categorical' : 
    {
        'Data' : read_file('2025_12_03/16/categorical/003'),
        'Color' : 'blue'
    },
    'Normal'      : 
    {
        'Data' : read_file('2025_12_03/16/normal/400x640/upscale2/008'),
        'Color' : 'orange'
    }
}

# Create 2x2 subplot grid
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Top-left: Fashion Accuracy
ax = axes[0, 0]
for dist, info in fashion_data.items():
    epoch = info['Data']['Training_Data']['Epoch']
    samples = epoch * 200 * 1000  # Assuming 200k samples per epoch
    rate = 1/300
    time = samples * rate # in seconds

    ax.plot(time, info['Data']['Training_Data']['Accuracy'], label=f'{dist} - Training', color=info['Color'])
    ax.plot(time, info['Data']['Validation_Data']['Accuracy'], label=f'{dist} - Validation', color=info['Color'],linestyle='--')

    # Draw star for test
    test_epoch = (info['Data']['Test_Data']['Epoch'].values[0] - 1) * 200 * 1000  # Assuming test is done after the last epoch
    test_time = test_epoch * rate # in seconds
    test_accuracy = info['Data']['Test_Data']['Accuracy'].values[0]
    ax.scatter(test_time, test_accuracy, marker='*', s=200, label=f'{dist} - Test',color=info['Color'])

ax.set_title('Accuracy')
ax.set_xlabel('Time [sec]')
ax.set_ylabel('Accuracy')
ax.grid()

legend = ax.legend(handles=[
    plt.Line2D([0], [0], color='blue', lw=2, label='Categorical'),
    plt.Line2D([0], [0], color='orange', lw=2, label='Normal'),
    plt.Line2D([0], [0], color='black', lw=2, label='Training'),
    plt.Line2D([0], [0], color='black', lw=2, linestyle='--', label='Validation'),
    plt.Line2D([0], [0], color='black', marker='*', linestyle='None', markersize=10, label='Test'),
], loc='lower right', bbox_to_anchor=(1, 0))

# Top-right: Fashion Loss
ax = axes[0, 1]
for dist, info in fashion_data.items():
    epoch = info['Data']['Training_Data']['Epoch']
    samples = epoch * 200 * 1000  # Assuming 200k samples per epoch
    rate = 1/300
    time = samples * rate # in seconds

    ax.plot(time, -info['Data']['Training_Data']['Loss'], label=f'{dist} - Training', color=info['Color'])
    ax.plot(time, -info['Data']['Validation_Data']['Loss'], label=f'{dist} - Validation', color=info['Color'],linestyle='--')
    
    # Draw star for test
    test_epoch = (info['Data']['Test_Data']['Epoch'].values[0] - 1) * 200 * 1000  # Assuming test is done after the last epoch
    test_time = test_epoch * rate # in seconds
    test_reward = -info['Data']['Test_Data']['Loss'].values[0]
    ax.scatter(test_time, test_reward, marker='*', s=200, label=f'{dist} - Test',color=info['Color'])

ax.set_title('Loss')
ax.set_xlabel('Time [sec]')
ax.set_ylabel('Loss')
ax.grid()

legend = ax.legend(handles=[
    plt.Line2D([0], [0], color='blue', lw=2, label='Categorical'),
    plt.Line2D([0], [0], color='orange', lw=2, label='Normal'),
    plt.Line2D([0], [0], color='black', lw=2, label='Training'),
    plt.Line2D([0], [0], color='black', lw=2, linestyle='--', label='Validation'),
    plt.Line2D([0], [0], color='black', marker='*', linestyle='None', markersize=10, label='Test'),
], loc='upper right', bbox_to_anchor=(1, 1))

# Bottom-left: MNIST Accuracy
ax = axes[1, 0]
for dist, info in mnist_data.items():
    epoch = info['Data']['Training_Data']['Epoch']
    samples = epoch * 200 * 1000  # Assuming 200k samples per epoch
    rate = 1/300
    time = samples * rate # in seconds

    ax.plot(time, info['Data']['Training_Data']['Accuracy'], label=f'{dist} - Training', color=info['Color'])
    ax.plot(time, info['Data']['Validation_Data']['Accuracy'], label=f'{dist} - Validation', color=info['Color'],linestyle='--')

    # Draw star for test
    test_epoch = (info['Data']['Test_Data']['Epoch'].values[0] - 1) * 200 * 1000  # Assuming test is done after the last epoch
    test_time = test_epoch * rate # in seconds
    test_accuracy = info['Data']['Test_Data']['Accuracy'].values[0]
    ax.scatter(test_time, test_accuracy, marker='*', s=200, label=f'{dist} - Test',color=info['Color'])

ax.set_title('Accuracy')
ax.set_xlabel('Time [sec]')
ax.set_ylabel('Accuracy')
ax.grid()

legend = ax.legend(handles=[
    plt.Line2D([0], [0], color='blue', lw=2, label='Categorical'),
    plt.Line2D([0], [0], color='orange', lw=2, label='Normal'),
    plt.Line2D([0], [0], color='black', lw=2, label='Training'),
    plt.Line2D([0], [0], color='black', lw=2, linestyle='--', label='Validation'),
    plt.Line2D([0], [0], color='black', marker='*', linestyle='None', markersize=10, label='Test'),
], loc='lower right', bbox_to_anchor=(1, 0))

# Bottom-right: MNIST Loss
ax = axes[1, 1]
for dist, info in mnist_data.items():
    epoch = info['Data']['Training_Data']['Epoch']
    samples = epoch * 200 * 1000  # Assuming 200k samples per epoch
    rate = 1/300
    time = samples * rate # in seconds

    ax.plot(time, -info['Data']['Training_Data']['Loss'], label=f'{dist} - Training', color=info['Color'])
    ax.plot(time, -info['Data']['Validation_Data']['Loss'], label=f'{dist} - Validation', color=info['Color'],linestyle='--')
    
    # Draw star for test
    test_epoch = (info['Data']['Test_Data']['Epoch'].values[0] - 1) * 200 * 1000  # Assuming test is done after the last epoch
    test_time = test_epoch * rate # in seconds
    test_reward = -info['Data']['Test_Data']['Loss'].values[0]
    ax.scatter(test_time, test_reward, marker='*', s=200, label=f'{dist} - Test',color=info['Color'])

ax.set_title('Loss')
ax.set_xlabel('Time [sec]')
ax.set_ylabel('Loss')
ax.grid()

legend = ax.legend(handles=[
    plt.Line2D([0], [0], color='blue', lw=2, label='Categorical'),
    plt.Line2D([0], [0], color='orange', lw=2, label='Normal'),
    plt.Line2D([0], [0], color='black', lw=2, label='Training'),
    plt.Line2D([0], [0], color='black', lw=2, linestyle='--', label='Validation'),
    plt.Line2D([0], [0], color='black', marker='*', linestyle='None', markersize=10, label='Test'),
], loc='upper right', bbox_to_anchor=(1, 1))

# Add dataset labels on the left side
fig.text(0.02, 0.75, 'Fashion MNIST', va='center', rotation='vertical', fontsize=14, fontweight='bold')
fig.text(0.02, 0.25, 'MNIST', va='center', rotation='vertical', fontsize=14, fontweight='bold')

plt.tight_layout(rect=[0.03, 0, 1, 1])
plt.show()
```
