```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from PythonTools.data_reader import read_file
```

```python
fashion_data = {
    'Categorical' : 
    {
        'Data' : read_file('2026_01_27/16/categorical/fashion/008_59_b_lr_0.15_sz_10'),
        'Color' : 'blue'
    },
    'Normal'      : 
    {
        'Data' : read_file('2026_01_27/16/normal/fashion/002_59_b_lr_0.15_sz_10'),
        'Color' : 'orange'
    }
}

# Plot the accuracy for both distributions
plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
for dist, info in fashion_data.items():
    epoch = info['Data']['Training_Data']['Epoch']
    samples = epoch * 200 * 1000  # Assuming 200k samples per epoch
    rate = 1/300
    time = samples * rate # in seconds

    plt.plot(time, info['Data']['Training_Data']['Accuracy'], label=f'{dist} - Training', color=info['Color'])
    plt.plot(time, info['Data']['Validation_Data']['Accuracy'], label=f'{dist} - Validation', color=info['Color'],linestyle='--')

    # Draw star for test
    test_epoch = (info['Data']['Test_Data']['Epoch'].values[0] - 1) * 200 * 1000  # Assuming test is done after the last epoch

    # Convert test epoch to time
    test_time = test_epoch * rate # in seconds
    
    test_accuracy = info['Data']['Test_Data']['Accuracy'].values[0]
    plt.scatter(test_time, test_accuracy, marker='*', s=200, label=f'{dist} - Test',color=info['Color'])

plt.title('Accuracy over Epochs for Categorical vs Normal Distribution')
plt.xlabel('Time [sec]')
plt.ylabel('Accuracy')
plt.grid()

legend = plt.legend(handles=[
    plt.Line2D([0], [0], color='blue', lw=2, label='Categorical'),
    plt.Line2D([0], [0], color='orange', lw=2, label='Normal'),
    plt.Line2D([0], [0], color='black', lw=2, label='Training'),
    plt.Line2D([0], [0], color='black', lw=2, linestyle='--', label='Validation'),
    plt.Line2D([0], [0], color='black', marker='*', linestyle='None', markersize=10, label='Test'),
], loc='lower right', bbox_to_anchor=(1, 0))
plt.gca().add_artist(legend)

plt.subplot(1,2,2)
# Plot the Loss
for dist, info in fashion_data.items():
    epoch = info['Data']['Training_Data']['Epoch']
    samples = epoch * 200 * 1000  # Assuming 200k samples per epoch
    rate = 1/300
    time = samples * rate # in seconds

    plt.plot(time, -info['Data']['Training_Data']['Loss'], label=f'{dist} - Training', color=info['Color'])
    plt.plot(time, -info['Data']['Validation_Data']['Loss'], label=f'{dist} - Validation', color=info['Color'],linestyle='--')
    
    # Draw star for test
    test_epoch = (info['Data']['Test_Data']['Epoch'].values[0] - 1) * 200 * 1000  # Assuming test is done after the last epoch
    test_time = test_epoch * rate # in seconds
    test_reward = -info['Data']['Test_Data']['Loss'].values[0]
    plt.scatter(test_time, test_reward, marker='*', s=200, label=f'{dist} - Test',color=info['Color'])

plt.title('Loss over Epochs for Categorical vs Normal Distribution')
plt.xlabel('Time [sec]')
plt.ylabel('Loss')
plt.grid()

legend = plt.legend(handles=[
    plt.Line2D([0], [0], color='blue', lw=2, label='Categorical'),
    plt.Line2D([0], [0], color='orange', lw=2, label='Normal'),
    plt.Line2D([0], [0], color='black', lw=2, label='Training'),
    plt.Line2D([0], [0], color='black', lw=2, linestyle='--', label='Validation'),
    plt.Line2D([0], [0], color='black', marker='*', linestyle='None', markersize=10, label='Test'),
], loc='upper right', bbox_to_anchor=(1, 1))
plt.gca().add_artist(legend)


plt.show()
```
